## Imports

```python
import tensorflow as tf
import matplotlib.pyplot as plt
print(tf.__version__)
```

## Data preparatie

```python
# Inlezen van de dataset
data_dir = 'archive/Dataset'
```

Instellen van de afbeeldingsdimensies en de batch size

```python
batch_size = 32
img_height = 180
img_width = 180
```

```python
train_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size)
```

```python
val_ds = tf.keras.utils.image_dataset_from_directory(
    data_dir,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(img_height, img_width),
    batch_size=batch_size,)
```

```python
class_names = train_ds.class_names
print(class_names)
```

### Data visualisatie

```python
plt.figure(figsize=(10, 10))
for images, labels in train_ds.take(1):
    for i in range(9):
        ax = plt.subplot(3, 3, i + 1)
        plt.imshow(images[i].numpy().astype("uint8"))
        plt.title(class_names[labels[i]])
        plt.axis("off")
```

## Model definitie & training

In deze sectie stel je het neural netwerk samen en wordt het getraind op de trainingsdata. Gebruik <code>model</code> als variabelenaam voor je model.

```python
# Vul dit zelf aan
num_classes = 5

model = tf.keras.Sequential([
    tf.keras.layers.Rescaling(1./255),
    #data_augmentation,
    
    #tf.keras.layers.Conv2D(32, 3),
    #tf.keras.layers.LeakyReLU(alpha=0.3),
    
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Conv2D(32, 3, activation='relu'),
    tf.keras.layers.MaxPooling2D(),
    tf.keras.layers.Dropout(0.2),
    
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(128, activation='relu'),
    tf.keras.layers.Dense(num_classes)
])
```

```python
model.compile(
    optimizer='adam',
    loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
    metrics=['accuracy'])
```

```python
model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=3
)
```

## Model evaluatie

```python
loss, acc = model.evaluate(val_ds)
print("Accuracy", acc)
```
